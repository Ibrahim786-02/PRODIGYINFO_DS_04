### Analyse and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.

```python
import pandas as pd
file_path = 'task4.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1', on_bad_lines='skip')
data.head(15)

```

```python
df = data.loc[:, ~data.columns.str.contains('^Unnamed')]

pd.DataFrame(df)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

positive_df = data[data['Sentiment'] == 'Positive']
negative_df = data[data['Sentiment'] == 'Negative']

positive_sentiment_counts = positive_df.shape[0]  # Count of positive sentiment rows
negative_sentiment_counts = negative_df.shape[0]  # Count of negative sentiment rows

plt.figure(figsize=(8, 6))
plt.bar(['Positive', 'Negative'], [positive_sentiment_counts, negative_sentiment_counts])  # Explicit labels and counts
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Positive vs. Negative Sentiment')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Assuming 'data' is the DataFrame containing the dataset

# Plot Sentiment Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Sentiment')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Plot Topic Sentiment Count
plt.figure(figsize=(14, 8))
sns.countplot(data=data, x='TOPIC', hue='Sentiment')
plt.title('Topic Sentiment Count')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()

# Optional: If you have a time column, you can plot sentiment over time
# Assuming you have a 'Time' column
# data['Time'] = pd.to_datetime(data['Time'])
# data.set_index('Time', inplace=True)
# data.groupby(['Time', 'Sentiment']).size().unstack().plot(kind='line', figsize=(14, 8))
# plt.title('Sentiment Over Time')
# plt.xlabel('Time')
# plt.ylabel('Count')
# plt.show()

# Bar Plot for Topics
plt.figure(figsize=(14, 8))
sns.countplot(data=data, x='TOPIC')
plt.title('Count of Tweets per Topic')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


```

```python
# Handle NaN values in the 'Tweet' column
data['Tweet'] = data['Tweet'].fillna('')

# Generate and Plot Word Cloud
all_tweets = ' '.join(data['Tweet'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_tweets)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Tweets')
plt.show()
```
